```python
# Suppress warnings
import warnings 
warnings.simplefilter('ignore')

import pyrcel as pm
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
P0 = 100000. # Pressure, Pa
T0 = 280.   # Temperature, K
S0 = -0.15  # Supersaturation, 1-RH (85% here)
```

```python
mu0 = 0.1
N0 = 50.0
kappa0 = 0.5

aer = pm.AerosolSpecies('aer', pm.Lognorm(mu=mu0, sigma=2.0, N=N0), kappa=kappa0, bins=50)
```

```python
initial_aerosols = [aer]
V = 1.5 # updraft speed, m/s

dt = 1.0 # timestep, seconds
t_end = 2000./V # end time, seconds... 2000 meter simulation 

model = pm.ParcelModel(initial_aerosols, V, T0, S0, P0, console=False, accom=0.3)
parcel_trace, aerosol_traces = model.run(t_end, dt, solver='cvode')
aer_array = aerosol_traces['aer'].values
```

```python
fig = plt.figure(figsize=(10,6))
fs = 12

endt = np.shape(aerosol_traces['aer'].values)[0]
tlist = np.linspace(0,endt,10,endpoint=False).astype(int)
#tlist = [0,20,25,30,35,40,45,50,80,110,140]

for t in tlist:
    r = aerosol_traces['aer'].values[t,:]
    plt.plot(r, aer.Nis*1e-6, linestyle='solid', marker='o', label="aerosol, t="+str(t))

plt.semilogx()
plt.xlabel("Aerosol dry radius ($\mu$m)",fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.ylabel("Aerosl number conc. (cm$^{-3})$",fontsize=fs)
plt.legend(loc='upper left',fontsize=fs)
plt.tight_layout()
plt.show()
```
